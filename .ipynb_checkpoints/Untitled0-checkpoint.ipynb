```python
import numpy
import scipy
from matplotlib import pyplot
from matplotlib import figure
import math
from scipy import special

N = 100;

# Part a: math.log(x) could be useful here.
'''
You could also do this part in WolframAlpha, but here is a setup for the Python code in case you would like to use it. You need to fill in the function to plot.

Plot ln(x) and save it in q6parta.png
'''

pyplot.clf()
pyplot.plot(range(1, N), [math.log(i) for i in range(1, N)]) # change me!
pyplot.xlabel("x")
pyplot.ylabel("ln(x)")
pyplot.savefig("q6parta.png", bbox_inches='tight', dpi=200)

# Part h
''' You could find math.sqrt(x), math.pi, math.e, and math.pow(x, p) to be useful

 Returns Stirling's Approximation to n! '''
def fact(n):
    k=1
    for i in range(1,n+1):
        k = k*i
    return k    

def choose(n,k):
    return fact(n)/((fact(n-k))*(fact(k)))

def stirling(n):
    c = math.pow(2*math.pi*n, 0.5)
    s = math.pow(n/math.e, n)
    return (c*s) # change me!

def f(n): 
   return stirling(n) / math.factorial(n)

# Plots f(n) and saves the plot to q6parth.png
pyplot.clf()
pyplot.plot(range(1, N), [f(i) for i in range(1, N)])
pyplot.xlabel("n")
pyplot.ylabel("f(n)")
pyplot.savefig("q6parth.png", bbox_inches='tight', dpi=200)


''' Part j: Define j_i(n) such that j_i(n) = ln( (n choose k_i) )
Note: math.log(x) and scipy.special.binom(n, k) could be useful here.
Keep in mind that math.log(x) takes the natural log of x.
'''
def j_1(n):
  return math.log(scipy.special.binom(n,0.25*n)) # change me!

def j_2(n):
  return math.log(scipy.special.binom(n,0.5*n)) # change me!

def j_3(n):
  return math.log(scipy.special.binom(n,0.75*n)) # change me!

# Plots ln((n choose k_i)) on the same plot and saves plot to q6partj.png
pyplot.clf()
pyplot.hold(True)
pyplot.plot(range(1, N), [j_1(i) for i in range(1, N)], color='red', label='m1 = 0.25')
pyplot.plot(range(1, N), [j_2(i) for i in range(1, N)], color='blue', label='m2 = 0.5')
pyplot.plot(range(1, N), [j_3(i) for i in range(1, N)], color='green', label='m3 = 0.75', linestyle='--')
pyplot.xlabel("n")
pyplot.ylabel("ln((n choose k))")
pyplot.legend()
pyplot.savefig("q6partj.png", bbox_inches='tight', dpi=200)

# Part k: Approximate the slope using a pair of points!

'''
TODO: Fill in slopes below. 
Use a point at n=70 and a point at n=90 to approximate the slopes of your lines.
Use these slopes for the rest of part k; putting this in this file was just 
to guide you on how you are expected to get them. Similar methods that give 
similar answers are also fine.
'''

# Slope of (n choose k_1)
slope1 = 0.4 # change me!

# Slope of (n choose k_2)
slope2 = 0.6 # change me!

# Slope of (n choose k_3)
slope3 = 0.4 # change me!

print("Slope 1: {0}, Slope 2: {1}, Slope 3: {2}".format(slope1, slope2, slope3))

def g_k1(n):
    return slope1*scipy.special.binom(n,0.25*n)

def g_k2(n):
    return slope1*scipy.special.binom(n,0.5*n)

def g_k3(n):
    return slope1*scipy.special.binom(n,0.75*n)
N = 1000
pyplot.clf()
pyplot.hold(True)
pyplot.plot(range(1, N), [g_k1(i) for i in range(1, N)], color='red', label='m1')
pyplot.plot(range(1, N), [g_k2(i) for i in range(1, N)], color='green', label='m2')
pyplot.plot(range(1, N), [g_k3(i) for i in range(1, N)], color='blue', label='m3')
pyplot.legend()
pyplot.savefig("q_last.png", bbox_inches='tight', dpi=200)
    
```
